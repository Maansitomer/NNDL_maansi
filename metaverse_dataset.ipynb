# ***Metaverse data in The Retail Industry***
The Study of Application of Metaverse in The Retail Industry And Its Impact on Consumer Buying Behavior to glean insights, we surveyed 400 frequent online shoppers, assessing their perceptions, expectations, and experiences related to Metaverse shopping. Alongside this, we also conducted an extensive literature review, analysing various academic articles, industry reports, and expert commentaries about the convergence of the Metaverse and retail.

**Basic Structure**
*   Total Rows: 400
*   Total Columns: 41

**Data Types:**
*   int64: 19 columns (Likert scale responses)
*   float64: 8 columns (Averages)
*   object: 14 columns (categorical/demographics)







```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
```

# Key Columns and Description

1.   **Demographic Information:**

      *   Email Address
      *   Age
      *   Gender
      *   Education
      *  Income
2.   **Measurement Sections:**
These Likert-scale questions (values: 1–5) are grouped and averaged.

      *   MA1–MA5: Metaverse Awareness → Average_MA
      *   MM1–MM5: Metaverse Motivation → Average_MM
      *   MC1–MC6: Metaverse Concerns → Average_MC
      *   PI1–PI3: Purchase Intention → Average_PI
3. **Product Categories:**

      *   Clothes, Accessories, Groceries, Electronics, Furniture, etc. (categorical preferences)





```python
df = pd.read_csv("Coded_File.csv")
df
```

```python
df = df.drop(columns=[col for col in df.columns if "Unnamed" in col or df[col].isna().all()])
```

```python
average_cols = ['Average_MA', 'Average_MM', 'Average_MC', 'Average_PI']
df[average_cols] = df[average_cols].apply(pd.to_numeric, errors='coerce')
```

```python
df.info()
```

```python
 df[average_cols].describe()
```

```python
print("\nCorrelation Matrix:\n", df[average_cols].corr())
sns.heatmap(df[average_cols].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Between Core Metrics")
plt.show()
```

**Inference**

* Motivation (MM) is the strongest predictor of purchase intention (PI).
* Awareness (MA) has a small but positive impact on both motivation and PI.
* Concerns (MC) have a minor inverse relationship with PI, meaning more concerns may lower purchase intent — though the effect is weak.
* These relationships suggest that boosting awareness and motivation may improve consumer adoption, while concerns do not drastically suppress interest but are worth monitoring.






```python
for col in average_cols:
    sns.histplot(df[col], kde=True, bins=20)
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

**Inferences:**

**Metaverse Awareness**

*   Slightly right-skewed bell curve.
*   Majority of values between 2.0 and 3.0.
*   Most respondents have low to moderate awareness of the Metaverse. There's a gradual buildup, indicating some familiarity, but not deep understanding or engagement.

**Metaverse Motivation**

*   Right-skewed but with a strong peak around 3.5–4.0.
*   A large portion of respondents show high motivation to use or explore the Metaverse despite low awareness. This suggests strong interest or curiosity even if understanding is limited.

**Metaverse Concerns**

*   Bimodal distribution — two visible peaks.Around 3.0 and 4.0
*   There is a clear divide in user sentiment. Some respondents have moderate concerns, while others express stronger concerns about using the Metaverse — possibly due to privacy, data, or usability issues. This bimodal nature suggests mixed opinions.

**Purchase Intention**

*   Discrete, possibly due to limited Likert scale options (1, 2, or 3).At values 2.0 and 3.0
*   Most respondents fall between neutral to slightly positive purchase intention. There’s no strong buying intent in the majority, which aligns with low awareness and high concern for some users.










```python
if "Gender" in df.columns:
    for col in average_cols:
        sns.boxplot(data=df, x="Gender", y=col)
        plt.title(f"{col} by Gender")
        plt.show()
```

**Inference:**

*   The gender-based behavior in the metaverse retail dataset is fairly balanced across the key metrics.
*   Slight tendencies:
    *   Males may show marginally higher values in Average_MM.
    *   Females slightly lead in Average_PI.
*   Outliers exist in both groups but are not extremely influential due to overall symmetric distributions.





```python
if "Age" in df.columns:
    grouped = df.groupby("Age")[average_cols].mean().reset_index()
    print("\nAverage scores by Age group:\n", grouped)
    grouped.plot(x="Age", kind="bar", figsize=(9,4))
    plt.title("Mean Scores by Age Group")
    plt.ylabel("Average Score")
    plt.xticks(rotation=45)
    plt.legend(title="Metrics")
    plt.tight_layout()
    plt.show()
```

**Inference:**

*   Primary Target: Age 21–30 — high awareness, marketing response, and usage.

*   Secondary Target: Age 31–40 — good conversion potential with enhanced awareness efforts.
*   Older Age Groups (41–60):
    *   Need educational or simplified onboarding strategies.
    *   Focus on trust, usability, and relevance to daily needs.
